#VISUALIZACIÓN

```python
#obtenemos la cantidad de universidades por pais
top = uni['Country'].value_counts().sort_values(ascending=False).head(5)
top
```

```python
#creamos un grafico de barras
bars = plt.bar(top.index, top.values, color = ['r', 'g', 'b', 'c', 'y'])

#ponemos un titulo a la grafica
plt.title('Top 5 de países con más universidades')

#Colocamos los valores en cada barra
for bar in bars:
  plt.text( bar.get_x() + bar.get_width()/2, bar.get_height(), bar.get_height(), ha= 'center', va = 'bottom')

#etiquetal para el eje x
plt.xlabel('PAÍSES')
#etiquetal para el eje y
plt.ylabel('CANTIDADES')
#Agregamos una cuadricula al grafico
plt.grid()

#Mostramos el grafico
plt.show()
```

```python
#Obtenemos los datos de por pais (Peru)
peru = uni[uni['Country'] == 'Peru']
```

```python
#Obtenemos las filas de la universidad especificada
cato = peru[ peru['Name'] == 'Pontifical Catholic University of Peru' ]
caye = peru[ peru['Name'] == 'Universidad Peruana Cayetano Heredia' ]
unmsm = peru[ peru['Name'] == 'Universidad Nacional Mayor de San Marcos' ]
upc = peru[ peru['Name'] == 'Universidad Peruana de Ciencias Aplicadas (UPC)' ]
```

```python
#obtenemos el grafico de cada universidad
plt.plot( cato['Year'], cato['Teaching'], color = 'orange', marker = 'o', label = 'Catolica')
plt.plot( caye['Year'], caye['Teaching'], color = 'blue', marker = '*', markersize = 8, label = 'Cayetano Heredia')
plt.plot( unmsm['Year'], unmsm['Teaching'], color = 'red', marker = 'D', label = 'San Marcos')
plt.plot( upc['Year'], upc['Teaching'], color = 'brown', marker = '^', label = 'UPC')

#Colocamos el titulo
plt.title('Calidad de enseñanza de 4 universidades del Perú (2018-2024)')

#Ponemos etiqueta al eje x
plt.xlabel('Años')

#Ponemos etiqueta al eje y
plt.ylabel('Puntuación sobre 100')

#Rotamos las etiquetas del eje x para un mejor legibilidad
plt.xticks(rotation=45)

#mostramos una leyenda
plt.legend()

#Mostramos una cuadricula en el grafico
plt.grid(linestyle='--')

#Mostramos el grafico
plt.show()
```

```python
#creamos una grafico con una tamaño especifico
plt.figure( figsize= (10,5) )
#Creamos una grafico de histogramsa
plt.hist(uni['Overall Score'], edgecolor = 'black', color = 'orange', bins = 20, alpha = 0.7)

#Colocamos un titulo
plt.title('Puntuación basada en la docencia, el entorno de investigación, la calidad de la investigación y las perspectivas internacionales')

#Mostramos una etiqueta para el eje x
plt.xlabel('Puntuación')

#Mostramos una etiqueta para el eje x
plt.ylabel('Frecuencia')

#Personalizamos las etiqueta del eje X
plt.xticks(np.arange(8, 100, 4.5), rotation = 45)

#Mostramos una cuadricula en el grafico
plt.grid()

#Mostramos el grafico
plt.show()
```

```python
#Creamos un df con los datos especificados ordenamos descendentemente
sca = uni[['Country', 'International Outlook', 'Year']].sort_values(by = 'International Outlook', ascending = False).head(5000)
sca2 = sca[sca['Year'] == 2024]
```

```python
#Creamos una grafica de mayor tamaño
plt.figure( figsize=  (20,10))

#Creamos un grafico de dispersion
plt.scatter(sca2['Country'],sca2['International Outlook'], label = 'CADA PUNTO ES UNA UNIVERSIDAD DISTINTA DE CADA PAIS')

#Rotamos las etiquetas del eje x para una mejor lectura
plt.xticks(rotation = 90)

#Colocamos un titulo
plt.title('Puntaje de las perspectivas internacionales de las universidades de cada pais (2024)')

#Colocamos una etiqueta en el eje x
plt.xlabel('PAÍSES')

#Colocamos una etiqueta en el eje y
plt.ylabel('PUNTUACIÓN')

#Mostramos la leyenda
plt.legend()

#Mostramos una cuadricula
plt.grid()

#Mostramos la grafica
plt.show()
```
